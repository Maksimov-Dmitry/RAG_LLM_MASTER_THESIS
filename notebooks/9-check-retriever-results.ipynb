```python
from aim import Repo

my_repo = Repo('../')

query = "run.experiment == 'predict_retriever'" # Example query
```

```python
metrics = my_repo.query_metrics(query)
```

```python
metrics = metrics.dataframe()
```

```python
metrics.columns
```

```python
metrics
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = metrics.loc[metrics['metric.name'].isin(['DocHitRate', 'HitRate'])]
```

```python
unique_metrics = df['metric.name'].unique()

# Set the overall figure size
plt.figure(figsize=(13, unique_metrics.size * 6))  # Adjust the size as needed

# Loop over each unique metric name and create a subplot
for i, metric in enumerate(unique_metrics, 1):
    plt.subplot(len(unique_metrics), 1, i)  # Creates subplot for each metric
    subset = df[df['metric.name'] == metric]
    
    # Create the barplot
    sns.barplot(
        data=subset,
        x='metric.context.subset',
        y='value',
        hue='run.hparams.model_name',
        dodge=True,  # This ensures bars for different hues don't overlap
         err_kws={'linewidth': 0}   # This removes the error bars
    )
    
    # Set the title for each subplot
    plt.title(f'Metric: {metric}')
    
    # Optional: Rotate the x-axis labels for better readability
    plt.xticks(rotation=45)

# Adjust the overall layout
plt.tight_layout()

# Show the plot
plt.show()
```
