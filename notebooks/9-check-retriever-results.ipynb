```python
metrics = metrics.dataframe()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = metrics.loc[metrics['metric.name'].isin(['DocHitRate', 'HitRate'])]
# df = metrics.loc[metrics['metric.name'].isin(['MRR@3'])]
```

```python
unique_metrics = df['metric.name'].unique()

# Set the overall figure size
plt.figure(figsize=(13, unique_metrics.size * 6))  # Adjust the size as needed

# Loop over each unique metric name and create a subplot
for i, metric in enumerate(unique_metrics, 1):
    plt.subplot(len(unique_metrics), 1, i)  # Creates subplot for each metric
    subset = df[df['metric.name'] == metric]
    
    # Ensure the 'step' column is treated as float
    subset['step'] = subset['step'].astype(int)

    # Create the lineplot
    sns.lineplot(
        data=subset,
        x='step',
        y='value',
    )
    
    # Set the title for each subplot
    plt.title(f'Metric: {metric}')
    # current_ticks = plt.xticks()[0]
    # plt.xticks(current_ticks, [int(tick) if tick.is_integer() else tick for tick in current_ticks])
    # Optional: Rotate the x-axis labels for better readability
    plt.xticks(rotation=45)

    # Add grid
    plt.grid(True)

# Adjust the overall layout
plt.tight_layout()

# Show the plot
plt.show()
```

```python
models = {
    'intfloat/multilingual-e5-large': 'Multilingual-e5-large',
    'models/retriever-e5-finetuned': 'Multilingual-e5-large, finetuned',
    'embed-multilingual-v3.0': 'Cohere-embed-multilingual-v3.0'
}
```

```python
df['run.hparams.model_name'].replace(models, inplace=True)
```

```python
unique_metrics = df['metric.name'].unique()

# Set the overall figure size
plt.figure(figsize=(13, unique_metrics.size * 6))  # Adjust the size as needed

# Loop over each unique metric name and create a subplot
for i, metric in enumerate(unique_metrics, 1):
    plt.subplot(len(unique_metrics), 1, i)  # Creates subplot for each metric
    subset = df[df['metric.name'] == metric]
    
    # Create the barplot
    sns.barplot(
        data=subset,
        x='metric.context.subset',
        y='value',
        hue='run.hparams.model_name',
        dodge=True,  # This ensures bars for different hues don't overlap
         err_kws={'linewidth': 0}   # This removes the error bars
    )
    
    # Set the title for each subplot
    plt.title(f'Metric: {metric}')
    
    # Optional: Rotate the x-axis labels for better readability
    plt.xticks(rotation=45)

# Adjust the overall layout
plt.tight_layout()

# Show the plot
plt.show()
```
