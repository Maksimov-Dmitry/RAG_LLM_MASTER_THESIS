```python
from transformers import AutoTokenizer, AutoModel
tokenizer = AutoTokenizer.from_pretrained('intfloat/multilingual-e5-large')
model = AutoModel.from_pretrained('intfloat/multilingual-e5-large')
```

```python
PREFIX = 'query: '
INPUT_MAX_LENGTH = 512
```

```python

def preprocess_data(examples):
  model_inputs = tokenizer(text=[PREFIX + text for text in examples['question']], max_length=INPUT_MAX_LENGTH, truncation=True, padding='max_length')
  return model_inputs
     
```

```python
dataset = dataset.map(preprocess_data, batched=True, num_proc=1)
```

```python
dataset.set_format(type='torch', columns=['input_ids', 'attention_mask'])
```

```python
from torch.utils.data import DataLoader
```

```python
dataloader = DataLoader(dataset['val'], batch_size=16, shuffle=False, drop_last=False, num_workers=5)
```

```python
import lightning as L
from torch import Tensor
import torch.nn.functional as F

class Model(L.LightningModule):
    def __init__(self, model_name):
        super().__init__()
        self.model = AutoModel.from_pretrained(model_name)

    def forward(self, input_ids, attention_mask):
        return self.model(input_ids, attention_mask=attention_mask)

    @staticmethod
    def average_pool(last_hidden_states: Tensor,
                    attention_mask: Tensor) -> Tensor:
        last_hidden = last_hidden_states.masked_fill(~attention_mask[..., None].bool(), 0.0)
        return last_hidden.sum(dim=1) / attention_mask.sum(dim=1)[..., None]

    def predict_step(self, batch, batch_idx, dataloader_idx=0):
        outputs = self.forward(**batch)
        embeddings = self.average_pool(outputs.last_hidden_state, batch['attention_mask'])
        embeddings = F.normalize(embeddings, p=2, dim=1)
        return embeddings
```

```python
model = Model('intfloat/multilingual-e5-large')
```

```python
trainer = L.Trainer()
```

```python
res = trainer.predict(model, dataloader)
```
